```python
def train(model, x, y, x_val, y_val, lr, epochs):
    log_dir = "logs/" + model.name + "/"
    
    tensorboard_callback = keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
    
    # look into using CSVlogger

    optimizer = tf.optimizers.Adam(learning_rate=lr)
    
    model.compile(
        optimizer=optimizer,
        loss=tf.keras.losses.MeanSquaredError(),
        metrics=["mean_squared_error"],
    )
    
    return model.fit(
        x, 
        y,
        epochs=epochs,
        validation_data=(x_val, y_val),
        callbacks=[tensorboard_callback],
        verbose=1,
    )
```

```python
simple_model = keras.models.Sequential(
    [
        Dense(2, input_dim=2, activation='sigmoid'),
        Dense(4, activation='relu'),
        Dense(1, kernel_initializer='normal')
    ]
)

simple_history = train(simple_model, x_train, y_train, x_val, y_val, 0.01, 10)

# run multiple times & avg.
```
